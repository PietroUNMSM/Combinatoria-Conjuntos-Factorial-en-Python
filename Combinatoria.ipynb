## SIN REPETICIÓN

```python
import math
```

```python
#VARIACIONES
def variacionSR(n,r):
    return math.factorial(n)/math.factorial(n-r)

n = int(input("Ingrese conjunto: ")) 
r = int(input("Ingrese subconjunto: "))

print('En variaciones sin repetición, para extraer ',r,' de ',n,' elementos tenemos ',int(variacionSR(n,r)), 'posibilidades')
```

```python
#COMBINACIONES
def numcombinaciones(n,k):           # CONDICION DEBE SER K <= N
    if(n<k):
        return 0
    else:
        combina= math.factorial(n)/(math.factorial(n-k)*math.factorial(k))
        return combina

conjunto = int(input("Ingrese conjunto: "))
subconjunto = int(input("Ingrese subconjunto: "))
print('En combinaciones sin repetición, sacando ',subconjunto ,' de ',conjunto,' elementos se tiene',int(numcombinaciones(conjunto,subconjunto)), 'casos posibles')
```

```python
#Variaciones
#VARIACIONES
def variacionCR(n,r):
    if (n<0) or (r<0):
        return 0
    else: 
        return math.pow(n,r)
    #return n**r

n = int(input("Ingrese conjunto: ")) 
r = int(input("Ingrese subconjunto: "))

print('En variaciones con repetición, para extraer ',r,' de ',n,' elementos tenemos ',int(variacionCR(n,r)), 'posibilidades')
```

```python
#PERMUTACIONES
#De cuántas maneras puedo distribuir 3 monedas de 50 centimos
# y 4 monedas de 20 centimos de tal forma que las 7 monedas queden alineadas??

n = int(input("Ingrese total de monedas: "))

r1 = int(input("Ingrese cantidad de monedas de 50 centimos: "))
r2 = int(input("Ingrese cantidad de monedas de 20 centimos: "))

def PermutacionCR(n,r1,r2):
    return math.factorial(n)/(math.factorial(r1)*math.factorial(r2))

print('En permutaciones con repetición, para distribuir',r1,'monedas de 50 centimos y',r2,
' monedas de 20 centimos tenemos',int(PermutacionCR(n,r1,r2)),'maneras') 
```

```python
#Combinaciones
import math 

def combinacionCR(n,r):
    if (n<r):
        return 0
    else:
        return math.factorial(n+r-1)/(math.factorial(n-1)*math.factorial(r))

n = int(input("Ingrese conjunto: ")) 
r = int(input("Ingrese subconjunto: "))

print('En combinaciones con repetición, para extraer ',r,' de ',n,' elementos tenemos ',int(combinacionCR(n,r))) 
    
```
